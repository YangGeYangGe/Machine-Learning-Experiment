```
def get_EMA(data,time_periods=12):
    a = 2/(time_periods+1)
#     for i in range(len(data)):
#         if i == 0:
#             data.loc[data.index[i], 'ema'] = data.loc[data.index[i], '<CLOSE>']
#         if i > 0:
# #             (Price [today] x K) + (EMA [yesterday] x (1 – K))
#             data.loc[data.index[i], 'ema'] = (a * data.loc[data.index[i], '<CLOSE>'])+(1-a)*data.loc[data.index[i] - 1, 'ema']
#     ema = list(data['ema'])
    #     return ema
  
    ema_list = []
    for i in range(len(data)):
        if i == 0:
            ema_list.append(data["<CLOSE>"].iloc[0])
        if i > 0:
            ema_list.append((a * data["<CLOSE>"].iloc[i] )+(1-a)*ema_list[i-1]) 
    return ema_list
```
